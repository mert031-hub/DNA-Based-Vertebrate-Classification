```python
plt.figure(figsize=(10,6))
sns.violinplot(x='target', y='gc_ratio', data=df)
plt.title("Türlere Göre GC Oranı Dağılımı")
plt.show()
```

```python
# Sadece ilk 20 k-mer ve biyometrikleri seçelim (tümü çok kalabalık olur)
#Hangi k-merlerin birbirine benzediğini veya hedef sınıfla (target) en çok hangi özelliğin ilişkili olduğunu gösterir.
# Mantık: Gereksiz (birbirinin aynısı olan) k-merleri tespit etmeni sağlar.
cols_to_plot = ['gc_ratio', 'entropy', 'cpg_ratio', 'mol_weight_avg'] + list(df.columns[:10])
plt.figure(figsize=(12,10))
sns.heatmap(df[cols_to_plot].corr(), annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Özellikler Arası Korelasyon Matrisi")
plt.show()
```
